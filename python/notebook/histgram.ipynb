```python
import numpy as np
import matplotlib.pyplot as plt 

bs = [
    '1B', '10B', '100B',
    '1KB', '10KB', '100KB',
    '1MB', '10MB', '100MB',
    '1GB', '10GB', '100GB',
]

with open('../data/datasize_units.txt', 'r') as input:
    ds = [l.strip() for l in input.readlines()]
    ls, cs = np.unique(ds, return_counts=True)
    ys = [0] * 12

    for l, c in zip(ls, cs):
        ys[bs.index(l)] = c
    ts = range(len(ys))

    plt.bar(bs, ys)
    plt.xticks(ts, bs)
```
